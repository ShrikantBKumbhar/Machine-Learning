```python
d1 = pd.DataFrame({'Key1':[10,11,12,13,14],
                   'Key2':[45,50,15,145,78],
                   'Key3':[786,10,111,130,141]})
```

```python
d1
```

```python
d2 = pd.DataFrame({'Key1':[15,41,82,11,15],
                   'Key5':[45,50,15,145,78],
                   'Key6':[786,10,111,130,141]})
```

```python
d2
```

```python
pd.merge(d1,d2) # merge is joinning two tables/df considering common records depending on common columns, by default it is inner join where output it only matching values from two table based on the commmon columns
```

```python
pd.merge(d1,d2,how = 'left') # left join - returns all the data from left table (d1) and only matching records from right table (d2)
```

```python
pd.merge(d1,d2,how = 'right') # right join - returns all the data from right table (d2) and only matching records from left table (d1)
```

```python
pd.merge(d1,d2,how = 'outer') # outer join return all matching and unmatching data from all tables, returns 1st columns all data
```

```python
pd.merge(d1,d2,how = 'cross') # it combines 1st table's 1st row with 2nd tables all rows and keep doing same with each row from first table till last row.
```

```python
pd.merge(d1,d2, how = "left", left_on="Key2", right_on = "Key5") # with this we can pass diff primary keys based on both tables to match the records based on these columns 
# this can be done in case where need to merge data of different columns/keys
```

```python
dd3 = pd.DataFrame({'Key1':[10,11,12,13,14],
                   'Key2':[45,50,15,145,78],
                   'Key3':[786,10,111,130,141]}, 
                  index = ['a','b','d','e','f']) # index will assign customized index values to rows
```

```python
dd4 = pd.DataFrame({'Key4':[15,41,82,11,15],
                   'Key5':[45,50,15,145,78],
                   'Key6':[786,10,111,130,141]},
                  index = ['a','b','g','h','i']) # index will assign customized index values to rows
```

```python
dd3
```

```python
dd4
```

```python
#JOIN in pandas
#join will merge tables based on index value as a key

dd3.join(dd4) # by default it takes left join, all records from left table and matching records from right table considering index as primary key
```

```python
dd3.join(dd4,how = 'right') # Returns all records from right table and matching records from left table considering index as primary key
```

```python
dd3.join(dd4,how = 'inner') # returns matching records from both the table considering index as primary key
```

```python
dd3.join(dd4,how = 'outer') # Returns all matching and unmatching records from both table considering index as primary key
```

```python
dd3.join(dd4,how = 'cross') # same like in merge, returns new index by default
```

```python
# Window functions

df6 = pd.DataFrame({'a': [1,2,3,4,5,6,7,8,9]})
```

```python
df6
```

```python
df6.rolling(window = 1).mean() #rolling roles to each row, window 1 means takes single no and calculate mean of each no - e.g. mean of 1 is 1 i.e. 1.0 and so on
```

```python
df6.rolling(window = 2).mean() #rolling roles to each row, window 2 means it takes two adjucent no and calculate mean of these two nos.
  # here 1st row shows NaN due to there is no any number before 1st place, so it takes 2 numbers e.g. from index 0 and 1 from original DF and calculates mean. e.g. 1+2/2 = 1.5
```

```python
df6.rolling(window = 3).mean() #rolling roles to each row, window 3 means takes 3 numbers and calculate mean of these 3 numbers
```

```python
df6.rolling(window = 3).sum()
```

```python
df6.rolling(window = 2).sum()
```

```python
df6.rolling(window = 3).min()
```
